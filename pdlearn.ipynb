````
df.City = df.City.str.strip()
df.City
````

````
# 去掉，
df.Price[100:200]
````

````
df.Price = df.Price.str.replace(',','')
````

````
# 转为int
import locale 
df.Price = df.Price.apply(locale.atoi)
df.Price 
````

````
df[df.Price > 3000]
````

````
# order by 
df.sort_values('Price')
````

````
df.sort_values('Price',ascending=False)
````

````
df.sort_values(['Price','City'], ascending=[False,True])
````

````
# group by 
df.groupby(df.Country)
````

````
# 分组之后取平均值，无字符串的字段无法取平均值，就消失了
df.groupby(df.Country).mean()
````

````
df.groupby(df.Country).mean().sort_values('Price', ascending=False)[:10]
````

````
# join on 
df2 = pd.DataFrame([['MO',0] ,['OR',1] ,['AL','2']],columns=['State','v'])
df2
````

````
# 默认为inner join 
pd.merge(df2,df, on='State')
````

````
pd.merge(df2,df, on='State', how='right')
````
