```
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```
x = np.arange(0.0, 2.0, 0.05) * np.pi
y1 = np.sin(x)
y2 = np.sin(2*x + np.pi)
```

```
plt.plot(x, y1)
```

```
plt.plot(x, y1, 'r', x, y2, 'b--')
plt.xlabel('this is x')
plt.ylabel('this is y')
plt.title('this is title')
plt.legend()
plt.plot()
```

```
plt.plot(x, y1, 'r' , label="y1")
plt.plot(x, y2, 'b--', label='y2')
plt.xlabel('this is x')
plt.ylabel('this is y')
plt.title('this is title')
plt.legend()
plt.plot()
```

## subplot 

```
plt.subplot(2,1,1) # 1st subplot
plt.plot(x,y1,'b--', label='y1')
plt.ylabel('this is y')
plt.subplot(2,1,2) # 2nd subplot
plt.plot(x,y2, 'r', label='y2')
plt.ylabel('this is y')
plt.xlabel('this is x')
```

### subplots

```
figure, ax = plt.subplots(2,2)
ax[0][0].plot(x,y1)
ax[0][1].plot(x,y2)
```

## Series 

```
s = Series(np.random.randn(1000))
```

```
s.plot()
```

```
s1 = s.cumsum()
```

```
s1.plot()
```

```
s1.plot(kind="line", grid=True, label='s1',title='Thist is title', style='--')
plt.legend()
```

```
s2 = Series(np.random.randn(1000)).cumsum()
```

```
s1.plot(kind="line", grid=True, label='s1',title='Thist is title', style='--')
s2.plot()
plt.legend()
```

```
fig, ax = plt.subplots(2,1)
s1.plot(ax=ax[0],label="s1")
s2.plot(ax=ax[1],label="s2")
plt.legend()
```
