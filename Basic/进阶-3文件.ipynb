linux有写缓存机制，只有在close()时才会完成写入。

1。 主动调用close(),flush(), 写缓存同步到磁盘。  
2。 写入数据量大于等于写缓存时，也会同步到磁盘。

### 文件指针
1。写入文件后，必须打开才能读取写入内容。  
2。读取文件后，必须重新再次读取读过的内容。   

文件指针的移动过程


os.SEEK_SET:0, start   
os.SEEK_CUR:1, current   
os.SEEK_END:2, end   

```python
f = open('1.txt','r+') # 必须先定义f, 才能在help中使用。
help(f.tell)
```

```python
f.tell()
```

```python
f.read(3) #执行一次走3位。
f.tell()
```

```python
import os 
f.seek(0,os.SEEK_SET) #回到文件首部，移动不能越界。
f.tell()
```

使用python 管理ini配置文件

```python
import configparser

cfg = configparser.ConfigParser()
```

```python
cfg.read('ini.txt')
```

```python
cfg.sections()
```

```python
for se in cfg.sections():
    print(se)
    print(cfg.items(se))
```

```python
cfg.set('userinfo','pwd','123456')
```

```python
for se in cfg.sections():
    print(se)
    print(cfg.items(se))
```

```python
cfg.set('userinfo','email','123456@qq.com')
for se in cfg.sections():
    print(se)
    print(cfg.items(se))
```

```python
cfg.remove_option('userinfo','email')
for se in cfg.sections():
    print(se)
    print(cfg.items(se))
```
