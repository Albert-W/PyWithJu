```
url = "http://samplecsvs.s3.amazonaws.com/SalesJan2009.csv"
df = pd.read_csv(url)
```

```
plt.plot(df.Transaction_date, df.Price)
```

```
plt.plot(df.Transaction_date.apply(str), df.Price)
```

```
plt.figure(figsize=(15,5))
plt.plot(df.Transaction_date, df.Price)
```

```
df2= df.sort_values("Transaction_date")
plt.figure(figsize=(15,5))
plt.plot(df2.Transaction_date, df2.Price)
```

```
# 先把str 转成 datetimelike values
# 然后按日期聚合
# 每隔一段时间显示x 
df3 = df.groupby([df['Transaction_date'].dt.date]).mean()
df3
```
